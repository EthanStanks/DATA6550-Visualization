### Import Libraries

```python
# Code Block 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_columns', 500)
plt.style.use('fivethirtyeight')
```

### Import Data

```python
# Code Block 2
df_gender_submission = pd.read_csv('/Users/jenner/Desktop/DATA 6550/data/gender_submission.csv')
display(df_gender_submission.head())
df_gender_submission.info()
```

```python
# Code Block 3
df_test = pd.read_csv('/Users/jenner/Desktop/DATA 6550/data/test.csv')
display(df_test.head())
df_test.info()
```

```python
# Code Block 4
df_train = pd.read_csv('/Users/jenner/Desktop/DATA 6550/data/train.csv')
display(df_train.head())
df_train.info()
```

```python
# Code Block 5
df_combined = pd.concat([df_train, df_test])
df_combined.info()
```

```python
# Code Block 6
df_clean = df_combined.dropna(subset = ['Survived', 'Embarked', 'Age', 'Fare', 'Sex'])
df_clean.info()
```

The datasets have been combined. Now, I will create the ethical graphs.

### Good Graphs

```python
# Code Block 7

survival_rate = (
    df_clean.groupby('Embarked')['Survived']
    .mean()
    .sort_index() * 100
)

port_labels = {
    'C': 'Cherbourg',
    'Q': 'Queenstown',
    'S': 'Southampton'
}

plt.figure()

plt.bar(
    [port_labels[p] for p in survival_rate.index],
    survival_rate.values, color = 'seagreen'
)

plt.xlabel("Embarked Port")
plt.ylabel("Survival Rate Percentage")
plt.title("Survival Rate by Embarked Port")
plt.ylim(0, 100)

plt.show()
```

```python
# Code Block 8
survived = df_clean[df_clean['Survived'] == 1]
died = df_clean[df_clean['Survived'] == 0]

plt.figure()

plt.scatter(died['Age'], died['Fare'], color = 'red', alpha = 0.5, label = 'Died')
plt.scatter(survived['Age'], survived['Fare'], color = 'green', alpha = 0.5, label = 'Survived')

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Fare vs Age by Survival")
plt.legend()

plt.show()
```

Now, I will create the unethical, very evil and mean spirited graphs.

### Evil Graphs

```python
# Code Block 9
survival = df_clean.groupby(['Sex', 'Survived']).size().unstack()

plt.figure()

plt.bar(
    survival.index,
    survival[1],
    color = 'red',
    label = 'Survived'
)

plt.bar(
    survival.index,
    survival[0],
    bottom = survival[1],
    color = 'green',
    label = 'Died'
)

plt.xlabel("Sex")
plt.ylabel("Number of Passengers")
plt.title("Survival by Sex")
plt.legend()

plt.show()
```

This is evil because the colors are swapped.
